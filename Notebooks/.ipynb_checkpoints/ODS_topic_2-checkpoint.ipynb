## TSNE метод

```python
from sklearn.manifold import TSNE
```

```python
tsne = TSNE(random_state=0)
```

```python
data_tsne = data.drop(['state','phone number'], axis=1)
```

```python
data_tsne['international plan']=data_tsne['international plan'].map({'yes':1,'no':0})
```

```python
data_tsne['voice mail plan']=data_tsne['voice mail plan'].map({'yes':1,'no':0})
```

```python
data_tsne.head()
```

```python
%%time

tsne.fit(data_tsne)
```

```python
plt.scatter(tsne.embedding_[:,0], tsne.embedding_[:,1]);
```

```python
import datetime as dt
```

```python
data_habr=pd.read_csv('../Datasets/howpop_train.csv.zip', compression='zip')
```

```python
data_habr.head(3).T
```

```python
data_habr.shape
```

```python
data_habr.drop([x for x in data_habr.columns if x.endswith('lognorm')], axis=1, inplace=True)
```

```python
data_habr['published'] = pd.to_datetime(data_habr.published, yearfirst = True)
```

```python
print(data_habr.published.dtype)
```

```python
data_habr['day']=pd.to_datetime(data_habr['published']).dt.date
```

```python
data_habr['year'] = [d.year for d in data_habr.published]
data_habr['month'] = [d.month for d in data_habr.published]
data_habr['dayofweek'] = [d.isoweekday() for d in data_habr.published]
data_habr['hour'] = [d.hour for d in data_habr.published]
```

```python
data_habr.head(3).T
```

```python
data_habr['year-month'] = data_habr['published'].apply(lambda x: str(x.year) + ' ' + str(x.month))
```

```python
data_habr.groupby('year-month', as_index=False).agg({'title':'count'}).sort_values(by='title', ascending=False).head()
```

```python
data_habr.query('year==2015 & month==3').pivot_table(index='day', values='title', 
                     aggfunc='count', columns='domain').plot(figsize=(12,5));
```

```python
data_habr.query('year==2015 & month==3')[data_habr.query('year==2015 & month==3').dayofweek == 6].pivot_table(index='day', 
                                              values='title', 
                                              aggfunc='count',
                                              columns='domain')
```

```python
data_habr.groupby('hour', as_index=False)[['views']].mean().sort_values('views', ascending=False).head()
```

```python
data_habr.groupby('hour', as_index=False)[['comments']].mean().sort_values('comments', ascending=False).head()
```
