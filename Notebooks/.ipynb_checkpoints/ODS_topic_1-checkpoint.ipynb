```python
#Найдите самое популярное имя среди пассажиров Титаника мужского пола?
```

```python
list_of_names = []

for y in titanic_data.Name:
    separated_name = y.split()
    
    for z in separated_name:
        list_of_names.append(z)

for z in list_of_names:
    reg = re.compile('[^a-zA-Z ]')
    cleared_name=reg.sub('', z)
    list_of_names.append(cleared_name)
```
